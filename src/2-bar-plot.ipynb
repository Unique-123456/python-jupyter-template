### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Error handling: Check if data arrays are valid
try:
    if len(categories) != len(values) or len(values) != len(colors):
        raise ValueError("Data arrays have mismatched lengths. Ensure categories, values, and colors have the same length.")
    
    # Plot
    plt.figure(figsize=(6, 4))
    plt.bar(categories, values, color=colors)

    # Customization
    plt.title("Bar Plot of Categories")
    plt.xlabel("Categories")
    plt.ylabel("Values")

    # Show plot
    plt.show()

except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

```
