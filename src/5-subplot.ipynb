### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
try:
    # Line plot data
    x = np.linspace(0, 10, 100)
    y = x**2

    # Bar chart data
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]

    # Scatter plot data
    scatter_x = [1, 2, 3, 4, 5]
    scatter_y = [2, 4, 6, 8, 10]

    # Pie chart data
    pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
    pie_sizes = [40, 25, 20, 15]

    # Check if data arrays are empty
    if len(x) == 0 or len(y) == 0 or len(categories) == 0 or len(values) == 0 or len(scatter_x) == 0 or len(scatter_y) == 0 or len(pie_labels) == 0 or len(pie_sizes) == 0:
        raise ValueError("One or more data arrays are empty. Please ensure all data is provided.")

    # Check if the length of pie sizes matches the labels
    if len(pie_sizes) != len(pie_labels):
        raise ValueError("Pie chart data arrays have mismatched lengths. Ensure labels and sizes have the same length.")

    # Create subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # Line plot
    axs[0, 0].plot(x, y, color='blue', label=r'$y = x^2$')
    axs[0, 0].set_title("Line Plot")
    axs[0, 0].set_xlabel("X values")
    axs[0, 0].set_ylabel("Y values")
    axs[0, 0].grid(True)
    axs[0, 0].legend()

    # Bar chart
    axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axs[0, 1].set_title("Bar Chart")
    axs[0, 1].set_xlabel("Categories")
    axs[0, 1].set_ylabel("Values")

    # Scatter plot
    axs[1, 0].scatter(scatter_x, scatter_y, color='blue')
    for i in range(len(scatter_x)):
        axs[1, 0].annotate(f"({scatter_x[i]}, {scatter_y[i]})", (scatter_x[i], scatter_y[i]), textcoords="offset points", xytext=(5,5), ha='center')
    axs[1, 0].set_title("Scatter Plot")
    axs[1, 0].set_xlabel("X values")
    axs[1, 0].set_ylabel("Y values")
    axs[1, 0].grid(True)

    # Pie chart
    axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, wedgeprops={'edgecolor': 'black'})
    axs[1, 1].set_title("Pie Chart")

    # Adjust layout
    plt.tight_layout()

    # Show plot
    plt.show()

except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
